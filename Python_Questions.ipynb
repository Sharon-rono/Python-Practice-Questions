```python
#Dictionaries
student = {"name": "Alice", "age": 21, "major": "Math"}
print(student["name"])

student["GPA"] = 3.8
print(student)

student["age"] = 22
print(student)

del student["major"]
print(student)

for key, value in student.items():
  print(f"{key} -> {value}")

fruits = ["apple", "banana", "apple", "cherry", "banana", "banana"]
counts = {}
for fruit in fruits:
  if fruit in counts:
    counts[fruit] += 1
  else:
    counts[fruit] = 1
print(counts)
```

```python
#Mixed Data Structures + Conditionals
inventory = {"apples": 5, "bananas": 2, "oranges": 0}
if inventory["oranges"] == 0:
  print("We are out of oranges")
else:
  print(f"We have {inventory["oranges"]} oranges")

people = [{"name":"Alice","age":30}, {"name":"Bob","age":17}]
for person in people:
  if person["age"] >= 18:
    print(f"{person["name"]} is an adult")
  else:
    print(f"{person["name"]} is a minor")

nums = (5, 10, 15, 20)
for num in nums:
  if num % 10 == 0:
    print(f"{num} is divisible by 10")
  else:
    print(f"{num} is not divisible by 10")

num1 = int(input("Enter First number"))
num2 = int(input("Enter Second number"))
num3 = int(input("Enter Third number"))
num_list = [num1, num2, num3]
num_set = set(num_list)
print(num_set)
if len(num_list) > len(num_set):
  print("The list had duplicates")
else:
  print("The list had no duplicates")

grades = {"Alice": 85, "Bob": 72, "Charlie": 90}
for name, grade in grades.items():
  if grade >= 80:
     print(f"{name} passed")
  else:
    print(f"{name} failed")

a = [1,2,3,4]
b = [3,4,5,6]
common = set(a).intersection(set(b))
print(common)

name = input("Enter Your name")
age = int(input("Enter your age"))
fruits = input("Enter your 3 favourite fruits separated by commas")
data = {"name": name, "age":age, "fruits":fruits}
print(f"Name:{data["name"]}, Age:{data["age"]}, Fruits: {data["fruits"]}")
```
